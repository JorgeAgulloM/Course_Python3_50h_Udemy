números, crea una tupla donde la primera entrada sea el número entero y, la segunda, la palabra “par” o

Dado un año proporcionado por el usuario, crea una tupla de dos elementos cuya primera entrada sea el año
y, la segunda entrada, el horóscopo chino correspondiente.


```python
#Se crea el diccionario con el calendario chino desde el 1960 a 2031
Chinese_horoscope = {
                    "Rata": [2020, 2008, 1996, 1984, 1972, 1960], 
                    "Buey": [2021, 2009, 1997, 1985, 1973, 1961],
                    "Tigre": [2022, 2010, 1998, 1986, 1974, 1962],
                    "Conejo": [2023, 2011, 1999, 1987, 1975, 1963],
                    "Dragón": [2024, 2012, 2000, 1988, 1976, 1964],
                    "Serpiente": [2025, 2013, 2001, 1989, 1977, 1965],
                    "Caballo": [2026, 2014, 2002, 1990, 1978, 1966],
                    "Cabra": [2027, 2015, 2003, 1991, 1979, 1967],
                    "Mono": [2028, 2016, 2004, 1992, 1980, 1968],
                    "Gallo": [2029, 2017, 2005, 1993, 1981, 1969],
                    "Perro": [2030, 2018, 2006, 1994, 1982, 1970],
                    "Cerdo": [2031, 2019, 2007, 1995, 1983, 1971]
                    }

#Se da indicaciones al usuario
print("Introduce un año desde el 1960 al 2031 para conocer el animal del horóscopo chino")
year = int(input("Año: "))

#Blucle que recorre los objetos del diccionario
for key, val in Chinese_horoscope.items():
    #Se comprueba si el año introducido está en el grupo
    if year in val:
        #Si lo está se muestra y finaliza el bucle
        print(year, key)
        break
```

Ejercicio 3

Dada una frase proporcionada por el usuario, crea una lista de tuplas indicando palabra, longitud de cada
palabra, letra inicial y posición que ocupan dentro de la frase.


```python
#Se pieda un frase y se corta
words = input("Introduce una frase: ")
words = words.lower()
words = words.split()

#Variable de control para la posición en la frase
i = 0
#Bucle que recorre las palabras de la frase
for word in words:
    #Se crea la tupla con la palabra, su longitud, la letra inicia y su posición en la frase 
    tupla = (word, len(word), word[0], i)
    #Se incrementa la variable y se muestra la tupla
    i += 1
    print(tupla)
```

Ejercicio 4

Haz que el usuario introduzca palabras hasta que introduzca una palabra vacía. Guarda todas las palabras
en una tupla y muestra la primera y la última introducidas haciendo uso del método unpacking.
PISTA: Para guardar los elementos de uno en uno vas a tener que utilizar un tipo de dato que no es tupla
y luego transformarlo a tupla.

```python
#Ss piede la frase y se crea una lista
words = str(input("Introduce palabras: "))
words_list = []

#Bucle condicionado a que el usuario devuelva una frase vacía
while words.isalpha():
    #Se añade la palabra a la lista y se pide una nueva
    words_list.append(words)
    words = str(input("Introduce palabras: "))

#Se recupera la primera y última palabrade la trasnformación de la lista en tupla
(word_start, *_, word_end) = tuple(words_list)

#Se muestra
print(word_start)
print(word_end)
```

Ejercicio 5

Dada una lista de palabras, crea otra lista del mismo tamaño que guarde la longitud de cada palabra. Usa
la función zip() para crear un diccionario con claves las palabras y valores, su longitud.


```python
#Lista de palabras y lista vacía
words = ["Python", "Programación", "Diversión", "futuro"]
len_words = []

#Bucle que recorre la lista de palabras y guarda la longitud de cada una en la lsita vacía
for word in words:
    len_words.append(len(word))

#Se crea un diccionario con las palabras como claves y las longitudes como valores
print(dict(zip(words, len_words)))
```

en una tupla y muestra la tupla y el número total de caracteres que ha introducido.
PISTA: Para guardar los elementos de uno en uno vas a tener que utilizar un tipo de dato que no es tupla
y luego transformarlo a tupla.


```python
#Se da instrucciones y se pide la primera palabra, se crea una lista vacía para alamcenarlas
words = input("Introduce palabras, usa una vacía para terminar: ")
list_words = []

#Bucle condicionado a que words sea alphanumérico
while words.isalpha():
    #Se agregan las palabras y se pide una nueva
    list_words.append(words)
    words = input("nueva palabra: ")

#Variable para contabilizar las letras
words_counter = 0
#Bucle que recorre las palabras guardadas
for word in list_words:
    #Bucle que recorre las letras de la palabras e incrementa en 1 el contador
    for c in word:
        words_counter += 1

tupla = tuple(list_words)

print(tupla, "Has introducido {} letras".format(words_counter))
```

muestra las tablas de multiplicar del 1 al 20 con el formato “1 x 1 = 1”.


```python
#Seimporta la libreria para generar un número random
import random as rd

#Se crea una lista y se rellena la primera entrada con un valor random
multiples = []
multiples.append(rd.randrange(1, 20))
#Se muestra
print("\n=== Número Random del 1 al 20 ===")
print(multiples)

#Se generan los múltiplos del número random
for i in range(0, 10 + 1):
    multiples.append(multiples[0] * i)

#Se muestran
print("\n=== Múltiplos de {} ===".format(multiples[0]))
print(multiples[1:11])

#Se generan las tablas de multiplicar del 1 al 20 y se agregan a la lista
for i in range(1, 20 + 1):
    #Variable para almacenar temporalmente una lista de la tabla de turno
    aux = []
    #Se muestra un encabezado para la tabla de turno
    print("\n=== Tabla del {} ===".format(i))
    for j in range(0, 10 + 1):
        #Se guarda el valor en la lista auxiliar y se muestra linea a linea
        aux.append("{} x {} = {}".format(i, j, i*j))
        print(aux[j])
    #Se guarda la lista auxiliar en la global
    multiples.append(aux)

#Se transforma el lista en una tupla y se muestra¿¿¿?????
tupla = tuple(multiples)
print("\n=== Tupla ===")
tupla
```

Realiza la división entera y guarda en una tupla el dividendo, el divisor, el cociente y el resto de la divisón
entera realizada y muéstrale al usuario el resultado por pantalla.


```python
#Se piden los dos números
n1 = int(input("Introduce el prinmer número: "))
n2 = int(input("Introduce el segundo número: "))

#Se comprueba que el primero sea igual o mayo que el segundo
if n1 >= n2:
    #Se calculan y guardan los valores
    tupla = (n1, n2, n1 // n2, n1 % n2)
    #Se muestra el resultado
    print(tupla)

#Si no se cumple la condición se muestra el mensaje al usuario
else:
    print("El primer número debe ser mayor o igual al segundo")
```

Ejercicio 9

Pide al usuario números entre 0 y 360. Para cada número, crea una tupla donde la primera entrada sea
dicho número y, la segunda, la medida angular correspondiente en radianes. Recuerda, 360◦ = 2πrad. En
este caso, utiliza π = 3.141592653589793


```python
#Se da indicaciónes y se pide el primer número al usuario
n = float(input("Introduce un número entre 0 y 360 (un número negativo termina el programa): "))

#Bucle condicionad a que no se introduzca un número negativo
while n > 0:
    #Se crea la tupla y se calculan guardando los datos
    tupla = (n, n * (2 * 3.141592653589793) / 360)
    #Se muestra el resultado y se pide un nuevo número
    print(tupla)
    n = float(input("Introduce otro número: "))
```
