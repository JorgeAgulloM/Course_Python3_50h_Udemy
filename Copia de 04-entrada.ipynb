# Filtrando solicitudes

Pyratilla hizo un formulario para que los interesados en formar parte de su tripulación rellenaran y así poder evitarse más de una entrevista innecesaria.

Ahora es momento de filtrar las convocatorias:

- El nombre debe empezar por mayúscula. Si es un nombre compuesto, entonces todos deben empezar por mayúscula.
- La edad debe ser mayor o igual a 16 y menor o igual a 40.
- El hobby indicado debe tener más de 10 caracteres.
- La casilla del sueño no puede haber sido dejada en blanco.

Ayuda a Pyratilla a crear este filtro para descartar solicitudes que no cumplan estos requisitos.

```
allOk = 0

#Nombre. El nombre debe comenzar con Mayússculas, si es compuesto el resto del nombre tambien.
name = input("Por favor, introduce tu nombre pirata: ")
if name.istitle():
    allOk += 1

#Edad. Debe ser mayor o igua a 16 hasta menor o igua a 40
age = int(input("Por favor, introduce edad: "))
if age >= 16 and age <= 40:
    allOk += 1

#Hobby. Debe ser un hobby con 10 caracteres como mínimo
hobby = input("Dime cual es tu hobby: ")
if len(hobby) >= 10:
    allOk += 1

#Sueño, no puede quedar en blanco
dream = input("Cual es el sueño de tu vida pirata: ")
if dream.isalpha:
    allOk += 1

if allOk == 4:
    print("El pirata '{}' con edad '{}' comprendida entre los 16 y los 40, \nal que le gusta el hobby de '{}' y tiene el sueño de '{}' \nserá llamado a la entrevista.".format(name, age, hobby, dream))
else:
    print("El pirata '{}' no pasa a la siguiente fase, gracias por inscribirte.".format(name))

```
