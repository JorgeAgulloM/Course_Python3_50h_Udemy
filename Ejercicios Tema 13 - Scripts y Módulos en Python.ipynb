```python
from vectors import Vector2D, Vector3D
```

Pruebas Vector 2D

```python
u = Vector2D(3, 4)
print(u)
```

```python
print("El módulo del vector u es =", u.module)
```

```python
u.scalar_prod(2)
print("2u = ", u)
u.scalar_prod(1 / 2)
```

```python
v = Vector2D(-1, 2)
print(v)
```

```python
print(u, "+", v, "=", Vector2D.sum(u, v))
```

```python
print(u, "-", v, "=", Vector2D.subtract(u, v))
```

```python
print("u * v =", u, "*", v, "=", Vector2D.dot_product(u,v))
```

```python
print("d(u, v) =", Vector2D.distance(u, v))
```

```python
u3d = u.extend_to_3d(5)
print(u3d)
```

```python
u = Vector3D(1, 0, -1)
print(u)
```

```python
print("u * v =", u, "*", v, "=", Vector3D.dot_product(u, v))
```

```python
print("d(u, v) =", Vector3D.distance(u, v))
```
