Ejercicio 1

Haz que el usuario introduzca números enteros por teclado. Mientras el usuario no introduzca el 0, muestra
si el número introducido es par o impar.


```python
#Se muestra un mensaje al usuario
print("Introduce números enteros, te diré si es par, cuando quieras terminar introduce el 0")
#Se declara la variable que almacenará el valor que introduzca el usuario
n = int(input("Introduce un número: "))
#Bucle while condicionado a que el usuario no introduzca el número 0
while n != 0:
    #El número introducido es...
    if n % 2 == 0: #... par
        print("El número {} es par.".format(n))
    else: #... impar
        print("El número {} es impar.".format(n))
    #Se pide un nuevo número al usuario
    n = int(input("Introduce un número: "))
    
#al salir del bucle se muestra un mensaje al usuario.
print("Gracias por participar.")
```

Ejercicio 2

Haz que el usuario introduzca una palabra y una letra por teclado. Comprueba si la palabra contiene la
letra o no e indícaselo al usuario por pantalla.


```python
#Se declaran la variables y se pide al usuario la entrada de datos
l = input("Por favor, introduce una palabra para examinarla: ")
l = l.lower()
c = input("Ahora indica una letra que quieras buscar en {}: ".format(l))
c = c.lower()

#Bucle for que recorre la cadena de la variable l hasta finalizar o encontrar la letra introducida en l
for i in range(len(l)):
    #Si la letra coincide con uno de los caracteres de la palabra
    if c == l[i]:
        #Se muestra al usuario y se sale del bucle
        print("La letra ({}) si está en la palabra ({})".format(c, l))
        break
#En caso de no salirdel bucle, se muestra al usuario un texto distinto indicando que no hay coincidencia
else:
    print("la palabra ({}) no contiene la letra ({})".format(l, c))
```

Ejercicio 3

Haz que el usuario introduzca precios por teclado (si introduce 0, entonces es que ha finalizado). Si el usuario
pasa de 200€, entonces ya no debe poder introducir más precios pues se ha pasado de presupuesto. Sea cual
sea el resultado (o bien el precio final o bien que no tiene más presupuesto), indícaselo por pantalla al usuario.

```python
#Se muestra las instrucciones al usuario
print("Introduce los precios que quieras, pero ojo, solo tienes 200€ de presupuesto. \nIntroduce 0 para finalizar la cuenta.")
#Se declaran las variables
p = 0
budget = 0

#Bucle while condicionado a que el presupuesto sea menos o igual a 200
while budget <= 200:
    #Se pide al usuario que introduzca un precio y se almacena
    p = float(input("intoduce un precio: "))
    #Se carga el valor acumulado y elnuevo introducido en una variable temporal
    budget_tmp = budget + p

    #Según el valor introducido...
    if p == 0:
        #... se notifica al usuario el presupuesto acumulado y se sale del programa
        print("Gracias por participar, tu presupuesto es de {}€".format(budget))
        break
    elif budget_tmp == 200:
        #... el usuario a conseguido el limite del presupuesto sin pasarse, se le notifica y se sale
        print("Muy bien, has conseguido el presupuesto máximo de 200€, adios.")
        break
    elif budget_tmp >= 200:
        #... el usuario se ha pasado del presupuesto, se le notifica y se sale
        print("Lo siento, te has pasado del presupuesto de 200€ en: {}€, adíos ".format(budget_tmp - 200))
        break
    else:
        #... el valor introducido cabe en el presupuesto, se notifica el nuevo presupuesto
        print("Llevas acumulados {}€".format(budget_tmp))
        #... y se carga el valor temporal en la variable global
        budget = budget_tmp
    
```

Ejercicio 4

Haz que el usuario introduzca números enteros por teclado. Mientras el usuario no introduzca el 0, calcula
cuántos números positivos y cuántos negativos ha introducido y muéstraselo al final.


```python
#Se muestra las instrucciones al usuario
print("introduce números enteros con el signo que quieras. Pulsa 0 para finalizar.")
#se pide la introducción de un valor
n = int(input("Introduce un número: "))
#Se declaran las variables para almacenar los número introducidos
count_n = 0
count_p = 0

#Bucle condicionado a que el usuairo no introduzca un 0
while n != 0:
    #Según el signo del número...
    if n < 0:
        #... se incrementa el contador de negativos
        count_n += 1
    else:
        #... se incrementa el contador de positivos
        count_p += 1
    
    #Se pide un nuevo número
    n = int(input("Introduce un número: "))

#Se muestra el resultado al usuario
print("Has introducido {} números positivos y {} números negativos.".format(count_p, count_n))
```

Haz que el usuario introduzca números por teclado. Mientras el usuario no introduzca el 0, pídele otro
número. Cuando el usuario introduzca el 0, muéstrale la media aritmética de los números que ha introducido.


```python
#Se muestra las instrucciones al usuario
n = float(input("Por favor, introduce un nñumero, (0 para finalizar): "))
#Se declaran variables para almacenar el valor de los número introducidos y su cantidad
sum = 0
count = 0

#Bucle condicionao que el usuario no introduzca un 0
while n != 0:
    #Se suma el nuevo valor introducido
    sum += n
    #Se incrementa el contador de número introducidos
    count += 1
    #Se pide un nuevo número
    n = float(input("Por favor, introduce un nñumero, (0 para finalizar): "))

#Se muestra el resultado
print("El valor médio aritmético de los valores introducidos es de: ", (sum / count))
```

Haz que el usuario introduzca dos números enteros por teclado. El primero será el extremo izquierdo del
intervalo y, el segundo, el extremo derecho. Imprime la suma de todos los múltiplos de 3 que se encuentren
entre los dos números introducidos por el usuario (los extremos incluidos). Finalmente, muestra por pantalla
el resultado de la suma.


```python
#Se pide al usuario que introduzca los valores
nL = int(input("Por favor, introduce un valor izquierdo para crear un intervalo: "))
nR = int(input("Ahora el valor Derecho para terminar el intervalo: "))
#Seimprime el primer número
print(nL)
#Variables para guardar el primer y último valor impreso
nLast = 0

#Bucle condicionado a que el valor izquierdo sea menor o igual al derecho
while nL <= nR:
    #Se incrementa el valor izquierdo para recorrer el intervalo
    nL += 1

    #Si el número es multiplo de 3 y no es le primer número impreso...
    if nL % 3 == 0:
        #... se imprime
        print(nL)
        #Se guarda el valor
        nLast = nL

#Si el último valor impreso no coincide con el fin de intervalo..
if nLast != nR:
    #... se imprime
    print(nR)
```

Ejercicio 8

Pídele al usuario cuántos números va a introducir. Con un bucle for, solicítale esa cantidad de números y
calcula su producto.

```python
#Se pide al usuario la cantidad de número que quiere introducir y se declaran las variables
c = int(input("Indica cuantos nñumero quieres introducir: "))
p = 0

#En el bucle for
for i in range(0, c):
    #Se pide al usuario que introduzca un número y se almacena temporalmente
    temp = float(input("Introduce el número {}: ". format(i + 1)))
    #Según el valor de p...
    if p == 0:
        #... se le carga el valor de temp
        p = temp
    else:
        #... se realiza el cálculo
        p = p * temp

#Se imprime el resultado
print(p)

```

Ejercicio 10

Haz que el usuario introduzca un número entero. Muestra un cuadrado y luego un triángulo rectángulo de
lado y altura, respectivamente, el número entero introducido. Por ejemplo, si el usuario introduce como
número 5, se deberá mostrar:
*            * * * * *
* *          * * * * *
* * *        * * * * *
* * * *      * * * * *
* * * * *    * * * * *

```python
#Se pide el número al usuario
n = int(input("Introduce un número: "))
#Se declara una variable para almacenar la línea
line = ""

#Bucle que recorre linea a linea para imprimir el resultado
for x in range(0, n):
    #Bucle que recorre la línea indicando que se debe añadir
    for z in range(0, (n * 2) + 1):
        #Se añade los primero puntos
        if z <= x:
            line = line + "*" #Añade el valor al string
        #Añade espacios
        elif z <= n:
            line = line + " "
        #Añade los puntos del final
        else:
            line = line + "*"
            break
    
    #Imprime el resultado y se borra la línea para volver a empezar.
    print(line)
    line = ""
```

Ejercicio de Pyratilla

99 bottles of rum on the wall

Pyratilla se pone a cantar:

"99 bottles of rum on the wall, 99 bottles of rum.
Take one down, pass it arraound, 98 bottles of rum on the wall.
98 bottles of rum on the wall, 98 bottles of rum.
Take one down, pass it arraound, 97 bottles of rum on the wall."

Ayuda a Pyratilla a terminar la canción, primero con un bucle while y despues sustituyelo por un bucle for.

Canción con bucle while

```python
#Entero que representa las iteraciones necesarias
i = 99
#Bucle condicionado a que la variable i sea mayor de 0
while i > 0:
    #Se imprime la primera frase
    print("{} bottles of rum on the wall, {} bottles of rum.".format(i, i))
    #Se reduce el indice
    i -= 1
    #Se imprime la segunda frase
    print("Take one down, pass it around, {} bottles of rum on the wall.".format(i))
```

Canción con bucle for

```python
#Bucle que recorre de 99 a 0
for i in range(99, 0, -1):
    #Seimprime la primera frase
    print("{} bottles of rum on the wall, {} bottles of rum.".format(i, i))
    #Se imprime la segunda frase, pero se reduce en uno el indice
    print("Take one down, pass it around, {} bottles of rum on the wall.".format(i - 1))
```
