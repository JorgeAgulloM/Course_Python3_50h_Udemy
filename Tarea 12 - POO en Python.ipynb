Ejercicio 10

Implementa el método de clase .toDate() que dado un string con el formato "01 January 0001" devuelva
el objeto fecha correspondiente a day = 1, month = 1, year = 1. Por ejemplo, si se pasa por parámetro
05 July 1985, el método debería devolver el objeto Date con los atributos day = 5, month = 7, year =
1985.


```python
class Date(object):

    def __init__(self, day = 1, month = 1, year = 1):
        self.day = day
        self.month = month
        self.year = year

        if self.validDay() == False:
            print("la fecha {}/{}/{} introducida no es valida.".format(self.day, self.month, self.year))

    def __str__(self):   
        return "{}/{}/{}".format(self.helperDate(1, self.day),
                                 self.helperDate(1, self.month),
                                 self.helperDate(2, self.year))

    def helperDate(self, object_date, value):
        """
        Devuelve una fecha con dos dijitos para dias, dos digitos para meses y
            vuatro digitos para años.
        Args:
            object_date (integer):
                1: para los días y los meses
                2: para los años
        Returns:
            (String) Fecha (día, mes o año) con formato 00/00/0000
        """
        str_value = str(value)
        while len(str_value) < object_date * 2:
            str_value = "0" + str_value
        return str_value

    def isLeap(self): 
        """
        Interpreta si el año que se ha introducido es bisiesto o no lo es
        Args:
            None
        Returns:
            Boolean con el resultado
        """
        if self.year % 4 == 0:
            return True
        return False
    
    def totalMonthDays(self):
        """
        Devuelve los días que teiene el mes correspondiente
        Args:
            None
        Returns:
            integuer con el número de días
        """
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif self.month == 2:
            if self.isLeap():
                return 29
            else:
                return 28
        return 30

    def validDay(self):
        """
        Verifica si el día introducido es correcto
        Args:
            None
        Returns:
            Boolean con la respuesta si es o no correcto

        """
        if self.year >= 0:
            if self.day > 0 and self.month > 0 and self.month <= 12:
                if self.day <= self.totalMonthDays():
                    return True
        return False

    @property
    def monthName(self):
        """
        Devuelve el nombre del mes introducido
        Args:
            None
        Returns:
            String con el nombre del mes
        """
        monthList = ["January", "February", "March", "April", "may", "June", 
                     "July", "August", "September", "October", "November", "December"]
        return monthList[self.month - 1]

    @staticmethod
    def areEqual(day1, month1, year1, day2, month2, year2):
        """
        Interpreta si las fechas introucidas son iguales
        Args:
            day1, day2 (integer): Valor de los días de cada fecha 
            month1, month2 (integer): Valor de los meses de cada fecha  
            year1, year2 (integer): Valor de los años de cada fecha 
        Resturns:
            Boolean con la respuesta
        """
        if day1 == day2 and month1 == month2 and year1 == year2:
            return True
        return False

    @staticmethod
    def areLater(day1, month1, year1, day2, month2, year2):
        """
        Interpreta si de las fechas introucidas la primera es mayor que 
            la segunda
        Args:
            day1, day2 (integer): Valor de los días de cada fecha 
            month1, month2 (integer): Valor de los meses de cada fecha  
            year1, year2 (integer): Valor de los años de cada fecha 
        Resturns:
            Boolean con la respuesta
        """
        if year1 > year2:
            return True
        elif year1 == year2:
            if month1 > month2:
                return True
            elif month1 == month2:
                if day1 > day2:
                    return True

        return False

    @staticmethod
    def arePrevious(day1, month1, year1, day2, month2, year2):
        """
        Interpreta si de las fechas introucidas la primera es menor que 
            la segunda
        Args:
            day1, day2 (integer): Valor de los días de cada fecha 
            month1, month2 (integer): Valor de los meses de cada fecha  
            year1, year2 (integer): Valor de los años de cada fecha 
        Resturns:
            Boolean con la respuesta
        """
        if year1 < year2:
            return True
        elif year1 == year2:
            if month1 < month2:
                return True
            elif month1 == month2:
                if day1 < day2:
                    return True

        return False

    @classmethod
    def firstDayOfTheYear(cls, year):
        """
        Devuelve un objeto Date con el primer día del año del año introducido
        Args:
            year (Integer): año a introducir
        Returns:
            Objeto Date
        """
        return Date(1, 1, year)

    @classmethod
    def lastDayOfTheYear(cls, year):
        """
        Devuelve un objeto Date con el último día del año del año introducido
        Args:
            year (Integer): año a introducir
        Returns:
            Objeto Date
        """
        return Date(31, 12, year)
    
    @classmethod
    def plusDay(cls, day, month, year):
        """
        Devuelve un objeto Date con el día siguiente a la fecha introducida
        Args:
            day (Integer): día a introducir
            month (Integer): mes a introducir
            year (Integer): año a introducir
        Returns:
            Objeto Date
        """
        upperDay = Date(day, month, year)   
        if upperDay.validDay(): 
            if day >= upperDay.totalMonthDays():
                if month == 12:
                    year += 1
                    month = 1
                else:
                    month += 1
                day = 1
            else:
                day += 1
            upperDay = Date(day, month, year)
            return upperDay

    @classmethod
    def minusDay(cls, day, month, year):
        """
        Devuelve un objeto Date con el día anterior a la fecha introducida
        Args:
            day (Integer): día a introducir
            month (Integer): mes a introducir
            year (Integer): año a introducir
        Returns:
            Objeto Date
        """
        lowerDay = Date(day, month, year)   
        if lowerDay.validDay(): 
            if day == 1:
                if month == 1:
                    year -= 1
                    month = 12
                else:
                    month -= 1
                    lowerDay.month -= 1
                day = lowerDay.totalMonthDays()
            else:
                day -= 1
            lowerDay = Date(day, month, year)
            return lowerDay

    @classmethod
    def copy(cls, date):
        """
        Devuelve una copia del objeto Date pasado por argumento
        Args:
            date (Objeto Date): objeto a copiar
        Returns:
            (Objeto Date): copia del objeto Date
        """
        import copy
        copy_date = copy.copy(date)
        return copy_date

    @staticmethod
    def difference(date1, date2):
        """
        Devuelve los días de diferencia que hay entre las fechas pasadas por 
            parametro
        Args:
            date1 (Objeto Date): fecha
            date2 (Objeto Date): fecha
        Returns:
            (Integer): Resultado de la diferencia entre fechas
        """
        date_low = date1
        date_higth = date2           

        if date1.year > date2.year:
            date_low = date2
            date_higth = date1
        elif date1.month > date2.month:
            date_low = date2
            date_higth = date1
        elif date1.day > date2.day:
            date_low = date2
            date_higth = date1

        day_diff = 0
        equal = False
        while equal == False:
            date_low = Date.plusDay(date_low.day, date_low.month, date_low.year)
            day_diff += 1
            if date_low.year == date_higth.year and date_low.month == date_higth.month and date_low.day == date_higth.day:
                equal = True

        return day_diff


    @classmethod
    def randomDate(cls):
        """
        Devuelve un Objeto Date random
        Args:
            None
        Returns:
            (Objeto Date) Objeto Date random valido
        """
        import random
        random_date = Date(random.randrange(32), random.randrange(13), random.randrange(3000))
        while random_date.validDay() == False:
            random_date = Date(random.randrange(32), random.randrange(13), random.randrange(3000))
        
        return random_date

    @classmethod
    def toDate(cls, dateString):
        """
        Devuelve la fecha formateada con dos dígitos el dia y el mes y 4
            dígitos para el año.
        Args:
            (String) Fecha en formato string
        Returns:
            (String) Fecha formateada
        """
        dateString = dateString.split(" ")
        monthList = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", 
                "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Dicimbre"]

        day = int(dateString[0])
        month = monthList.index(dateString[1]) + 1
        year = int(dateString[2])

        date = Date(day, month, year)
        return date
```
