
Ejercicio 1

Dado un número entero introducido por teclado, guarda sus divisores en un conjunto y muéstralo

```python
#Se pide un número
n = int(input("Introduce un número entero: "))
#Se declara un conjunto de datos vacío
divs = set()

if n >= 0:
    #Bucle que recorre n iteraciones
    for d in range(1, n):
        #Si el número tiene resto 0 significa que el valor d es divisor de n y se agrega al conjunto
        if n % d == 0:
            divs.add(d)

    #Se muestran los valores.
    print("Los divisores del número {} son {}".format(n, divs))
else:
    print("No has escrito un número positivo, vuelve a intentarlo.")
```

Ejercicio 2

Crea un programa que dado un conjunto, nos devuelva su mínimo. Debes hacerlo sin recurrir a la función
min().


```python
#Se declara un conjunto de dats y una variable entera con un número alto.
set1 = {3, 5, 9, -4, -25, 20}
min = 999999

#Bucle que recorre los items de set1
for n in set1:
    #Si el valor del item es menor que el de min se guarda en esta última
    if n < min:
        min = n

#Se muestra el resultado
print("el número mínimo del conjunto es", min)
```

Ejercicio 3

Dada una frase introducida por teclado, guarda en un conjunto todas las palabras que empiecen por la letra
indicada por el usuario.

```python
#Se pide la frase, que se corta con la función split, y la letra
s = input("Escribe una frase: ")
s = s.lower()
s = s.split()
c = input("Escribe una letra: ")

#Se de clara un conjunto de datos vacío
setWords = set()

#Bucle recorre las palabras almacenadas
for i in range(len(s)):
    #si la primera letra de cada palabra coincide con la letra que indica el usuario
    if s[i][0] == c:
        #Se agrega la palabra al conjunto
        setWords.add(s[i])

#Se muestra el conjunto
setWords
```

Ejercicio 4

Dado un conjunto, crea un programa que nos devuelva el caracter con mayor valor ASCII. Debes hacerlo sin
recurrir a la función max().


```python
#Conjunto de datos y valiables para guardar el número y la letra más altas en tabla ASCII
s = {"s", "o", "r", "H", "p", "@"} ##NOTA para hacerlo con una cadena con número, hay que pasar los valores por str()
value = 0
l = ""

#Bucle que recorre el conjunto
for c in s:
    #Si la posición en la tabla es mayor se guarda dicha posición en la variable
    if ord(c) > value:
        value = ord(c)
        #Se guarda la letra, que será la última que se guarde 
        l = c

#Se muestra el resultado
print("La letra {} con la posición ASCII {} es la alta".format(l, value))
```

Ejercicio 10

Dada una frase introducida por teclado, guarda en un conjunto todas las palabras palíndromas.

```python
#Se pide una frase con palabras palíndormas y se corta la frase pos sus espacios con split()
s = input("Intrduce una frase que incluya palabras palídromas: ")
s = s.split()
#Se declara un conjunto de datos vacío y una boolean para marcar las palabras palíndromas
setP = set()
palindrome = False

#Se recorren las palabras de s
for c in s:
    #Variable para almacenar la palabra al reves
    c_rev = ""
    
    #Se recorre letra a letra y se añaden a la inversa en la variable
    for j in range(len(c)):
        c_rev = c[j] + c_rev
    
    #Se recorre la palabra original y la invertida
    for i in range(len(c)):
        #Se comparan las letras y si coinciden se activa la bool
        if c[i] == c_rev[i]:
            palindrome = True
        else: 
            #... si no coincide una de ellas se pasa a False la bool y se sale del bucle
            palindrome = False
            break
    
    #Si la palabra es palindroma se añade al conjunto de datos
    if palindrome:
        setP.add(c)

#Se muestra el conjunto
setP
```
