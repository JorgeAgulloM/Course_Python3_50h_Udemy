Ejercicio 1

Dibuja 100 puntos aleatorios (entre 0 y 100 para ambas coordenadas) con colores y tamaños aleatorios y una
transparencia de 0.5.
Tendrás que crear 4 arrays aleatorios de 100 posiciones: uno para las coordenadas x, otro para y, otro para
el color de cada punto y un último para el tamaño de cada punto. Para que los puntos no se vean muy
pequeños, multiplica el array de tamaños por 10.
Configura el tamaño de la figure a 10 x 10. Para ello, investiga el método .figure() de plt.
Por último, selecciona los colores combinando el array aleatorio y el mapa de color llamado hsv. Recuerda
mostrar la colorbar.

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import seaborn as sns
```

```python
x_points = []
y_points = []
sices = []
colors = []
color = sns.color_palette("hsv")

for _ in range(100):
    x_points.append(np.random.randint(100))
    y_points.append(np.random.randint(100))
    colors.append(color[np.random.randint(len(color))])
    sices.append(np.random.randint(100) * 10)

ej1 = plt.scatter(x_points, y_points, c = colors, s = sices, alpha = 0.5)
ej1 = plt.colorbar()
ej1 = plt.figure(figsize= (10, 10))
ej1.show
```

Ejercicio 2

En el plano complejo, representa los números complejos 1 + i, 2i, −1 − i, −3, −2 − i, −2i y 1 − i.
Recuerda que el plano complejo es como un plano cartesiano, salvo que el eje horizontal se denomina eje real
y, el eje vertical, eje imaginario.
Representa los puntos de color rojo y tamaño 120 y los ejes vertical y horizontal de color negro y transparencia
0.5. Para representar los puntos, puedes considerar una lista que contenga las partes reales y otra lista que
contenga las partes imaginarias de los números complejos. Para representar los ejes, debes dibujar una recta
horizontal y = 0 y otra vertical x = 0. Investiga para ello el método .linspace() de numpy. Para que las
dimensiones del argumento x e y sean las mismas, puede que tengas que usar el método .zeros de numpy
para obtener un array de ceros del tamaño deseado.
Pon como título “Plano complejo” e investiga los métodos .xlim e .ylim de plt para establecer el rango de
x en el intervalo [-3.2, 3.2] y el rango de y en el intervalo [-1.2, 1.2].


```python
num = [(1 + 1j), (1 + 2j), (-1 -1j), (-3 + 0j), (-2 -1j), (1 -2j), (1 -1j)]
x = []
y = []
for i in num:
    x.append(i.real)
    y.append(i.imag)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.title("Plano complejo")
plt.scatter(x, y, c = "r", s = 120, alpha = 0.5)
plt.xlim(-3.2, 3.2)
plt.ylim(-1.2, 1.2)
plt.show
```

```python
real = [1, 2, -1, -3, -2, -2, 1]
imaginary = [1, 0, -1, 0, -1, 0, -1]
plt.title("Plano complejo")
x = np.linspace(-3.5, 3.5, 100)
plt.plot(x, np.zeros(100), c = "k", alpha = 0.5)
y = np.linspace(-1.5, 1.5, 100)
plt.plot(np.zeros(100), y, c = "k", alpha = 0.5)
plt.scatter(x = real, y = imaginary, c = "r", s = 120, alpha = 1)
plt.xlim([-3.2, 3.2])
plt.ylim([-1.2, 1.2])
plt.show()
```

Ejercicio 7

Dibuja el siguiente grafo con los nodos de color #ad80fa, las aristas de color #b1ddf5, el nombre de los
nodos de color blanco y en negrita.

```python
import networkx as nx
G = nx.Graph()
G.add_nodes_from("abcde")
G.add_weighted_edges_from([("a", "b", 1), ("a", "c", 2), ("a", "d", 2), ("a", "e", 5),
("b", "e", 4), ("c", "d", 3), ("d", "e", 1)])

nx.draw(G, with_labels = True, font_color = "w", font_weight = "bold", node_color = "#ad80fa", edge_color = "#b1ddf5")
```

Ejercicio 8

Considera el grafo del ejercicio anterior. El tamaño de los nodos debe ser 350 por el grado del nodo. El
tamaño de la figure global debe ser de 8 x 6.


```python
plt.figure(figsize= (8, 6))
nx.draw(G, with_labels = True, font_color = "w", node_size = [350 * G.degree(x) for x in G.nodes], font_weight = "bold", node_color = "#ad80fa", edge_color = "#b1ddf5")
plt.show
```

```python
nx.draw(G, with_labels = True, font_weight = "bold", font_color = "white",
node_color = "#ad80fa", edge_color = "#b1ddf5",
node_size = [350 * G.degree(x) for x in G.nodes])
plt.show()
```

Ejercicio 9

Considera el grafo del ejercicio anterior. El grosor de las aristas debe ir acorde al peso de cada arista.


```python
nx.draw(G, with_labels = True, font_color = "w", node_size = (350 * len(G.nodes)), font_weight = "bold", node_color = "#ad80fa", edge_color = "#b1ddf5", 
        width = (
            G.edges["a", "b"]["weight"], 
            G.edges["a", "c"]["weight"],
            G.edges["a", "d"]["weight"], 
            G.edges["a", "e"]["weight"],
            G.edges["b", "e"]["weight"],
            G.edges["c", "d"]["weight"],
            G.edges["d", "e"]["weight"]
            ))
plt.show
```

```python
plt.figure(figsize = (8, 6))
nx.draw(G, with_labels = True, font_weight = "bold", font_color = "white",
node_color = "#ad80fa", edge_color = "#b1ddf5",
node_size = [350 * G.degree(x) for x in G.nodes],
width = [G[u][v]['weight'] for u,v in G.edges])
plt.show()
```

Ejercicio 10

Investiga el método .text() de plt para añadir texto a los plots.
Añade las siguientes 3 palabras “LA”, “AVENTURA”, “SIGUE”.
La palabra “LA” debe estar en la posición (0.2, 0.8), con tamaño 50 y rotación 25, con alineación vertical y
horizontal en el centro, y en una caja de estilo redondo, con color de borde y relleno #242fff y transparencia
0.7.
La palabra “AVENTURA” debe estar en la posición (0.5, 0.55), con tamaño 50 y rotación -15, con alineación
vertical y horizontal en el centro, y en una caja de estilo redondo, con color de borde y relleno #8a24ff y
transparencia 0.7.
La palabra “SIGUE” debe estar en la posición (0.8, 0.3), con tamaño 50 y rotación 10, con alineación vertical
y horizontal en el centro, y en una caja de estilo redondo, con color de borde y relleno #24fff4 y transparencia
0.7.
El tamaño de la figure global debe ser 10 x 8.


```python
plt.figure(figsize= (10, 8))
plt.text(
    0.2, 0.8, s = "LA", 
    size = 50,
    rotation = 25,
    verticalalignment ='center', 
    horizontalalignment ='center',
    bbox ={'facecolor':'#242fff', 'pad':10, 'boxstyle':"round, pad=0.5, rounding_size=1",},
    alpha = 0.7)

plt.text(
    0.5, 0.55, s = "AVENTURA", 
    size = 50,
    rotation = -15,
    verticalalignment ='center', 
    horizontalalignment ='center',
    bbox ={'facecolor':'#8a24ff', 'pad':10, 'boxstyle':"round, pad=0.5, rounding_size=1",},
    alpha = 0.7)

plt.text(
    0.8, 0.3, s = "SIGUE", 
    size = 50,
    rotation = 10,
    verticalalignment ='center', 
    horizontalalignment ='center',
    bbox ={'facecolor':'#24fff4', 'pad':10, 'boxstyle':"round, pad=0.5, rounding_size=1",},
    alpha = 0.7)

plt.show
```
