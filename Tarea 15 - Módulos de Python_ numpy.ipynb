Dado el array 2D [["ala", "delfín", "arroz"], ["cena", "kayak", "picnic"], ["hoja", "gato",
"elle"]], busca las palabras que sean palíndromos.
PISTA: Necesitarás una función universal.


```python
def palindorme(array):
    """
    Devuelve las palabras palindromas pasadas por parametro en un array
    Args:
        array (numpy.array 2D)
    Returns:
        array (numpy.array) con los valores palíndormos
    """
    result = []
    for array_l2 in array:
        for word in array_l2:
            isPalindrome = False
            for i in range(len(word) // 2):
                if word[i] == word[-1 - i]:
                    isPalindrome = True
                else:
                    break
            if isPalindrome == True:
                result.append(word)

    return print(result)

a = np.array([["ala", "delfín", "arroz"], ["cena", "kayak", "picnic"], ["hoja", "gato", "elle"]])

myPalindormeFunct = np.frompyfunc(palindorme, 1, 1)
palindorme(a)
```

Ejercicio 9

Crea una función universal que dado un array de strings devuelva el número total de vocales de cada string.


```python
def total_vowels(array):
    """
    Devuelve el número total de vocales de cada string del array 
    pasado por parámetro.
    """
    a_str = np.array(array)
    result = 0
    for i in a_str.reshape(-1):
        for j in i:
            if j.lower() in ["a", "e", "i", "o", "u"]:
                result += 1
    
    return result


total_vowels = np.frompyfunc(total_vowels, 1, 1)

a = np.array([["ala", "delfín", "arroz"], ["cena", "kayak", "picnic"], ["hoja", "gato", "elle"]])
print(total_vowels(a))
```
