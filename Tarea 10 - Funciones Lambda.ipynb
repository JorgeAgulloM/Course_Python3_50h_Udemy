Ejercicio 3

Dada una lista de palabras, quédate con filter() con las que tengan más vocales que consonantes. Necesitarás una función que devuelva si una palabra tiene más vocales que consonantes.

```python
def consonants_vowels(word):
    """
    Devuelve las consonantes y las vocales que hay en la palabra dada por
    parámetro
    Args:
        word (String): palabra a examinar
    Returns:
            consonant_count (Integer): con el número de consonantes
            vowels_count (Integer): con el número de vocales
    """
    consonant_count = 0
    vowels_count = 0
    for c in word.lower():
        if c in ["a", "e", "i", "o", "u"]: #if c == "a" or c == "e" or c == "i" or c == "o" or c == "u":
            vowels_count += 1
        else:
            consonant_count += 1
    if vowels_count > consonant_count:
        return True
    return False
```

```python
words = ["tutu", "jugueteria", "violonchelo", "fanfarrón", "piano", "murcielago"]
list(filter(consonants_vowels, words)) #list(filter(lambda w:  consonants_vowels(w), words))
```

Ejercicio 4

Dada una lista de números enteros, quédate con filter() con los que tengan más de 5 divisores. Necesitarás
una función que devuelva el número de divisores de un número dado.

```python
def num_dividers(n):
    """
    Devuelve el número de divisores que tiene un número
    Args:
        n (Integer): número a examinar
    Returns:
        dividers_count (Integer): con la cantidad de divisores del número
    """
    dividers_count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            dividers_count += 1
        
    return dividers_count
```

```python
nums = [8, 12, 60, 14, 30, 22]
list(filter(lambda n: num_dividers(n) > 5, nums))
```

Ejercicio 5

Dada una lista de palabras, quédate con reduce() con la palabra más larga. Necesitarás una función que
compare dos palabras y devuelva la que tenga mayor longitud.

```python
def higher_word(w1, w2):
    """
    Devuelve la palabra con más caracteres
    Args:
        w1 (String): palabra 1 a comparar
        w2 (String): palabra 2 a comparar
    Returns:
        return (String): La palabra resultado 
    """ 
    if len(w1) >= len(w2):
        return w1
    return w2
```

```python
from functools import reduce

words = ["tutu", "jugueteria", "violonchelo", "fanfarrón", "piano", "murcielago"]
reduce(higher_word, words) #reduce(lambda w1, w2: higher_word(w1, w2), words)
```

Ejercicio 6

Dada una lista de palabras, calcula el número de vocales de cada una con map().

```python
def vowels_counter(word):
    """
    Devuelve la cantidad de vocales que hay en la palabra dada por
    parámetro
    Args:
        word (String): palabra a examinar
    Returns:
        vowels_count (Integer): cantidad de vocales 
    """
    vowels_count = 0
    for c in word.lower():
        if c in ["a", "e", "i", "o", "u"]: #if c == "a" or c == "e" or c == "i" or c == "o" or c == "u":
            vowels_count += 1
    
    return vowels_count
```

```python
words = ["tutu", "jugueteria", "violonchelo", "fanfarrón", "piano", "murcielago"]
list(map(lambda w: vowels_counter(w) , words))
```
