Ejercicio 1

Crea un programa que pida un número entero positivo por teclado y que cree un diccionario cuyas claves
sean desde el número 1 hasta el número indicado. Los valores de cada clave serán las propias claves elevadas
al cubo.


```python
#Se solicita un número y se declara el diccionario
n = int(input("Introduce un número: "))
dicc = {}

#Bucle que corre hasta n iteraciones y carga la clave y su valor al diccionario
for i in range(1, n):
    dicc[i] = i ** 3

#Se muestra
print(dicc)
```

Ejercicio 3

Escribe un programa que cree un diccionario simulando una cesta de la compra. El programa debe preguntar
el artículo y su precio por unidad. El artículo será la clave y el valor el precio, hasta que el usuario decida
terminar. Después se debe mostrar por pantalla la lista de la compra y el coste total, con el siguiente formato

Artículo 1 Precio

Artículo 2 Precio

Artículo 3 Precio

. . . . . .

Total Precio Total


```python
#Se da instrucciones al usuario y se declara el diccionario
print("Escribe los articulos de la cesta de la compra y su precio. \nIntroduce 0 en el articulo cuando quieras terminar")
list_buy = {}

#Se inicia con el primer artículo
n = str(input("Artículo: "))

#Bucle condicionado a que el usuario no introduzca el valor 0 (cero)
while n != "0":
    #Se carga el articulo como clave y se solicita el precio o valor
    list_buy[n] = float(input("Precio: "))
    #Se solicita el nuevo artículo (clave)
    n = str(input("Artículo: "))

#Variable para almacenar el valor totalde los artículos
valueT = 0
print("\n=== Lista de la compra ===")

#Bucle que recorre el diccionario guardando ambos valores por cada item
for article, value in list_buy.items():
    #Se muestran los art´culos introducidos y se calcula el valor total
    print("{}: {}€".format(article, value))
    valueT += value

#Se muestra el total
print("\tTotal: ", (valueT))
```

Ejercicio 4

Crea un programa que lea números enteros hasta que introduzca el 0 y devuelva un diccionario con la
cantidad números positivos y negativos introducidos.


```python
#Se establece la variable entera a 1 para poder iniciar el bucle 
n = 1
#Se crea una lista
numbers = [0, 0]

#Bucle condicionado a que el usuario no introduzca 0
while n != 0:
    #Se solicita al usuario un número
    n = int(input("Introduce número enteros positivos o negativos de uno en uno: "))
    #Según su signo se incrementa un contador...
    if n > 0:
        numbers[0] += 1
    elif n < 0:
        numbers[1] += 1
    
#Se crea y completa el diccionario
dicc_numbers = {"Positivos": numbers[0], "Negativos": numbers[1]}

#Se muestra la colección
print("\n=== Cantidad de enteros ===")
for num in dicc_numbers.items():
    print(num)
```

Ejercicio 5

Crea un programa que lea números enteros hasta que introduzca el 0 y devuelva un diccionario con la
cantidad números pares e impares introducidos.

```python
#Se establece la variable entera a 1 para poder iniciar el bucle 
n = 1
#Se crea una lista
numbers = [0, 0]

#Bucle condicionado a que el usuario no introduzca 0
while n != 0:
    #Se solicita al usuario un número
    n = int(input("Introduce número enteros de uno en uno: "))

    #Se interrumpe el bucle en caso de que se introduzca el 0 para que no lo cuente como nº par
    if n == 0:
        break

    #Según su signo se incrementa un contador...
    if n % 2 == 0:
        numbers[0] += 1
    else:
        numbers[1] += 1

#Se crea y completa el diccionario
dicc_numbers = {"Pares": numbers[0], "Impares": numbers[1]}

#Se muestra la colección
print("\n=== Cantidad de enteros ===")
for num in dicc_numbers.items():
    print(num)
```

Ejercicio 6

Crea un programa que permita al usuario introducir los nombres de los alumnos de una clase y las notas que
han obtenido. Cada alumno puede tener distinta cantidad de notas. Guarda la información en un diccionario
cuyas claves serán los nombres de los alumnos y los valores serán listas con las notas de cada alumno.

```python
#Se pide al usuario que indique cuntos alumnos va a ingresar
n_students = int(input("¿Cuantos alumnos vas a introducir?: "))
#Se declara un diccionario vacío
students = {}

#bucle que crea tantas posiciones como alumnos haya indicado el usuario
for i in range(n_students):
    #Se declara una lista temporal y un entero para almacenar nota a nota y controlar el bucle
    values = []
    v = 0

    #Se pide al usuario que intoduzca los nombres de los alumnos
    n = str(input("Nombre del alumno: "))
    #Bucle para agergar notas, se termina al introducir un número negativo
    while v > -1:
        v = float(input("Introduce una nota: "))
        if v >= 0: #Esto evita que se añada el valor de nota negativo para la salida
            values.append(v)

    #Se crea la nueva clave con el nombre y los valores con las notas, y se pide un nuevo nombre    
    students[n] = values

#Se muetra los pares del diccionario
print("\n=== Lista de alumnos y sus notas ===")
for student in students:
    print(student, ":", students[student], "- Nota media =", sum(students[student]) / len(students[student]))
```

Ejercicio 7

Crea un programa que pida un número entero positivo por teclado y que cree un diccionario cuyas claves
sean desde el número 1 hasta el número indicado. Los valores de cada clave serán tantos símbolos "*" como
indique la clave.

```python
#Se pide al usuario el número clave y se declara una diccionario vacío
n = int(input("Escribe un número entero: "))
dic = {}

#Bucle que recorre tantas iteraciones como haya indicado el usuario
for i in range(1, n + 1):
    #Variable temporal par guardar los valores de cada iteración del siguiente bucle
    values = ""
    #Bucle que genera los valores * según el indice i
    for j in range(i):
        values = values + "*"
    #Se carga la clave y los valores
    dic[i] = values

#Se muestra todo
for val in dic.items():
    print(val)
```

Ejercicio 8

Crea un programa que pida el número de palabras a introducir. Crear un diccionario de clave la palabra y
de valor la longitud de dicha palabra.

```python
#Se pide al usuario que indique cuntos alumnos va a ingresar
n_words = int(input("¿Cuantos palabras vas a introducir? "))
#Se declara un diccionario vacío
words = {}

#bucle que crea tantas posiciones como alumnos haya indicado el usuario
for i in range(n_words):
    #Se pide al usuario que intoduzca los nombres de los alumnos
    n = str(input("Introduce la palabra: "))

    #Se crea la nueva clave con el nombre y los valores con las notas, y se pide un nuevo nombre    
    words[n] = len(n)

#Se muetra los pares del diccionario
print("\n=== Lista de valores ===")
for word in words:
    print(word, ":", words[word])
```

Ejercicio 9

Crea un programa que pida el número de palabras a introducir. Crear un diccionario de clave la palabra y
de valor el número de vocales de la palabra.


```python
#Se pide al usuario que indique cuntos alumnos va a ingresar
n_words = int(input("¿Cuantos palabras vas a introducir? "))
#Se declara un diccionario vacío
words = {}

#bucle que crea tantas posiciones como alumnos haya indicado el usuario
for i in range(n_words):
    #Se declara una lista temporal y un entero para almacenar nota a nota y controlar el bucle
    values = 0

    #Se pide al usuario que intoduzca los nombres de los alumnos
    n = str(input("Introduce la palabra: "))

    for c in range(len(n)):
        if n[c] == "a" or n[c] == "e" or n[c] == "i" or n[c] == "o" or n[c] == "u":
            values += 1
        

    #Se crea la nueva clave con el nombre y los valores con las notas, y se pide un nuevo nombre    
    words[n] = values

#Se muetra los pares del diccionario
print("\n=== Lista de valores ===")
for word in words:
    print(word, ":", words[word])
```

Ejercicio 10

Dada una matriz, crea un diccionario que guarde el número de filas, el de columnas y cada fila en una entrada
de un diccionario.

```python
#Se carga una matriz y se declara un diccionario vacío
matrix = [[0, 1, 2], [3, 4, 5], [6, 7, 8]]
dicc_matrix = {}

#Se alamcenan los valores de filas y columnas
dicc_matrix["Filas"] = len(matrix)
dicc_matrix["Columnas"] = len(matrix[0])

#Bucle que recorre tantas iteraciones como filas haya
for i in range(len(matrix)):
    #Se guarda la fila en el diccionario
    dicc_matrix["Fila {}".format(i)] = matrix[i]

#Se muestran los datos formateados
print("=== esta es la Matriz cargada ===")
for fila in dicc_matrix:
    print(fila, ":", dicc_matrix[fila])
```
