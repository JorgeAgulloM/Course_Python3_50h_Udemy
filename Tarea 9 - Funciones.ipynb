Ejercicio 1

Crea una función que busque todos los divisores del número entero positivo dado por parámetro y devuelva
una lista con todos los divisores de dicho número.

```python
def divisors(n):
    """
    Función que busca los divisores del número entero positivo indicado por parámetro.
    Args:
        n (Integuer): número entero positivo.
    Returns:
        return: divisores del número indicado.
    """

    divisors_list = []
    for i in range(1, n + 1):
        if n % i == 0:
            divisors_list.append(i)
    
    return print("los divisores de {} son {}.".format(n, divisors_list))   
```

```python
divisors(20)
```

Crea una función que dados dos números reales por parámetro, devuelve el mayor.


```python
def higher(x, y):
    """
    Función que devuelve el mayor de los número reales pasados por parametro.
    Args:
        x: Número real a indicar
        y: Número real a indicar
    Retunrs:
        return: Número más alto de los dos.
    """
    if x == y:
        return print("Ambos números son del mismo tamaño.")
    elif x < y:
        t = y
        y = x
        x = t

    #return print("{} es mayor que {}.".format(x, y))
    return x
```

Ejercicio 4

Crea una función que devuelva True si el caracter introducido por parámetro se trata de una vocal y False
en caso contrario.

```python
def consonant_vowel(letter):
    """
    Función que evalua el caracter pasado por parámetro, si es vocal devolverá
        True, si es consonante devolverá False
    Args:
        letter (String): Letra a evaluar
    Returns:
        return: valor boolean (True para vocal, False para consonate)
    """
    if type(letter) != type("a"):
        return print("La función solo admite carácteres en forma de String")
    letter = letter.lower()
    if letter == "a" or letter == "e" or letter == "i" or letter == "o" or letter == "u":
        return True
    else:
        return False
```

```python
print(consonant_vowel(1))
print(consonant_vowel("A"))
print(consonant_vowel("b"))
```

Ejercicio 5

Crea una función que devuelva el MCD (máximo común divisor) de 2 números proporcionados por parámetro.
PISTA: Aprovecha la función que calcula el mayor entre dos números dados. También necesitarás una función
que calcula el menor entre dos números dados.

```python
def helper_factorice(x):
    """
    Función que realiza la factorización de un númer, pasado por parámetro, 
    que devuelve una lista con los factores primos de dicho número.
    Args:
        x (integer): Número a factorizar
    Returns:
        return: lista con los factores
    """
    factors = []
    f = 2
    while x > 1:
        if x % f == 0:
            x = x // f
            factors.append(f)
        else:
            f += 1

    return factors

def MCD(x, y):
    """
    Función que devuelve el Máximo Común Divisor de los número enteros 
        proporcionados por parámetros.
    Args:
        x (Integer): Número entero a evaluar
        y (Integer): Número entero a evaluar
    Returns:
        return: MCD = resultado de MCD(x, y)
    """
    t = higher(x, y)
    if t == y:
        y = x
        x = t
    
    factorice_x = helper_factorice(x)
    factorice_y = helper_factorice(y)
    
    if len(factorice_x) >= len(factorice_y):
        f1 = factorice_y
        f2 = factorice_x
    else:
        f1 = factorice_x
        f2 = factorice_y
        
    common_factors = []
    for i in f1:
        for j in f2:
            if j == i:
                common_factors.append(j)
                break

    MCD = 1
    for i in common_factors:
        MCD *= i

    return MCD 
```

```python
print(MCD(72, 50))
```

Ejercicio 6

Crea una función que devuelva el MCM (mínimo común múltiplo) de 2 números proporcionados por
parámetro.
PISTA: Aprovecha la función que calcula el MCD de dos números del ejercicio anterior y la función que
calcula el valor absoluto de un número.

```python
def MCM(x, y):
    """
    Función que devuelve el mínimo comun multiplo de dos número. Esta función
    tambien usa la función absolute(x) para hayar previemente el valor 
    absoluto del los número indicados y despues el función MCD(x, y) para
    obtener el Máximo Comun Divisor.
    Args:
        x (Integuer): Número a evaluar
        y (Integuer): Número a evaluar
    Returns:
        return: devuelve el mcm de los valores dados
    """
    x = absolute(x)
    y = absolute(y)

    return (x * y) // MCD(x, y)
```

Ejercicio 7

Crea una función que dada una palabra devuelva si es palíndroma.


```python
def palindrome(word):
    """
    Devuelve True o False según sea o no palíndroma la palabra pasada por 
    parámetro.
    Args:
        word (String): palabra a examinar.
    Returns:
        return: True o False si es o no palindroma.
    """
    word = word.lower()
    is_palindrome = False
    for c in range(len(word) // 2):
        if word[c] == word[-1]:
            is_palindrome = True
        else:
            is_palindrome = False
            break
    
    return is_palindrome
```

```python
palindrome("ojos")
```

Crea una función que dado un color en hexadecimal devuelva una lista de 3 posiciones, cada una de ellas
correspondiente al valor R, G o B en este orden. Los valores de RGB varían entre 0 y 255.


```python
def RGB(color):
    """
    Devuelve por separado los valores R, G y B en una lista dado el valor por 
    parámetro en exadecimal
    Args:
        color (String): e. g. #ff5566 valor del color a devolver
    Returns:
        return: Lista de los valores en tipo entero [R, G, B]
    """
    R, G, B = 0, 0 ,0
    if len(color) == 7:
        R = int(color[1:3], 16)
        G = int(color[3:5], 16)
        B = int(color[5:], 16)
        
        return [R, G, B]
    else:
        print("El valor que has introducido no es correcto.")
```

Ejercicio 9

Crea una función que dada una lista de palabras por parámetro, devuelva un diccionario que contenga
cuántas son de longitud par y cuántas de longitud impar.

```python
def dict_par_impar(sentence):
    """
    Devuelve un diccionario con las palabras introducidas en la frase por 
    parámetro y claseficadas entre par e impar por su longitud
    Args:
        sentence (String): Frase a revisar
    Returns:
        return: dict{} con las palabras clasificadas entre par o impar
    """
    sentence = sentence.split(" ")

    dict_pi = {"impar": [], "par": []}
    for word in sentence:
        if len(word) % 2 == 0:
            dict_pi["impar"].append(word)
        else:
            dict_pi["par"].append(word)

    return dict_pi
```

```python
dict_par_impar("Estoy a punto de comer")
```

Ejercicio 10

Crea una función que dado un string por parámetro cuente cuántas veces sale cada caracter en dicho string
y devuelva toda esa información en un diccionario.

```python
def count_char(sentence):
    """
    Cuenta los caracteres del string dado por parámetro y los lista en un 
    diccionario con los caracteres como vocales y el número de veces
    que aparece como valores.
    Args:
        sentece (String): Frase a evaluar
    Resturns:
        return: Dict{} con los caracteres como claves y las veces que 
                aparecen como valores
    """
    sentence = sentence.lower()
    sentence.split()

    dict_chars = {}
    counted = []
    for c in sentence:
        if c not in counted:
            dict_chars[c.upper()] = sentence.count(c)

    return dict_chars
```
